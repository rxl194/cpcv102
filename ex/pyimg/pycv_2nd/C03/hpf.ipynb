```python
# %load hpf.py
import cv2
import numpy as np
from scipy import ndimage
from matplotlib import pyplot as plot
%matplotlib inline

kernel_3x3 = np.array([[-1, -1, -1],
                   [-1,  8, -1],
                   [-1, -1, -1]])

kernel_5x5 = np.array([[-1, -1, -1, -1, -1],
                       [-1,  1,  2,  1, -1],
                       [-1,  2,  4,  2, -1],
                       [-1,  1,  2,  1, -1],
                       [-1, -1, -1, -1, -1]])

img = cv2.imread("../images/statue_small.jpg", 0)

k3 = ndimage.convolve(img, kernel_3x3)
k5 = ndimage.convolve(img, kernel_5x5)

blurred = cv2.GaussianBlur(img, (17,17), 0)
g_hpf = img - blurred

plot.subplot(221), plot.imshow(k3), plot.title("3x3")
plot.subplot(222), plot.imshow(k5), plot.title("5x5")
plot.subplot(223), plot.imshow(g_hpf), plot.title("g_hpf")
```
